### 租房資料清整_2

```python
import numpy as np
import pandas as pd
```

```python
rent = pd.read_csv('台北5.csv')
rent
```

```python
F=rent.租賃層次.str.split('([\u4e00|\u4e8c|\u4e09|\u56db|\u4e94|\u516d|\u4e03|\u516b|\u4e5d|\u5341]+)(\u5c64)').str.get(1)
rent['floor']=F
rent.head(5)
```

```python
# 抓出樓層非空值
rent2=rent[(rent['floor'].isnull()!=True)]
rent2.head(5)
```

```python
from cn2an import cn2an
# 把非空值樓層轉換成數字，再寫回原 dataframe 新欄位
rent['F']=rent2.loc[:,'floor'].apply(lambda x: str(cn2an(x,"normal")))
rent
```

```python
rent3=rent[(rent['floor'].isnull()!=True)]
rent4=rent3[(rent3['F'].isnull()!=True)]
rent5=rent4[(rent4['總樓層數'].isnull()!=True)]
rent5
```

```python
# 把土地 建物 車位 分別取出

L=rent5['租賃筆棟數'].str.split('(\u571f\u5730)(\d+)(\u5efa\u7269)(\d+)(\u8eca\u4f4d)(\d+)').str.get(2)
B=rent5['租賃筆棟數'].str.split('(\u571f\u5730)(\d+)(\u5efa\u7269)(\d+)(\u8eca\u4f4d)(\d+)').str.get(4)
P=rent5['租賃筆棟數'].str.split('(\u571f\u5730)(\d+)(\u5efa\u7269)(\d+)(\u8eca\u4f4d)(\d+)').str.get(6)
rent5['Land']=L
rent5['Building']=B
rent5['Parking_lot']=P

rent5
```

```python
# 換單位 平方公尺換成坪

rent5['單價元坪'] =rent5['單價元平方公尺'].apply(lambda x: x*3.31 )
rent5['建物總坪數'] =rent5['建物總面積平方公尺'].apply(lambda x: x*0.303 )
rent5
```

```python
# 將租賃年月日的年分出來
rent5['租賃年月日']=rent5['租賃年月日'].astype(str)
Y=rent5.租賃年月日.str.split('(10[0-9])').str.get(1)
rent5['Year']=Y
rent5
```

```python
# 找出空值
rent6=rent5[(rent5['Year'].isnull()==True)]
rent6
```

```python
# 由於上述空值並非在住宅分析範圍內，直接刪除
rent7 = rent5.drop([29856, 32182, 49295])
rent7
```

```python
# 把字串改為整數型態
rent7['Year']=rent7['Year'].astype(int)
rent7['總樓層數']=rent7['總樓層數'].astype(int)
rent7
```

```python
rent7.rename(columns={'Ads':'Address'}, inplace = True)
rent7.head(5)
```

```python
import re
house_type = []
list1 = rent7['建物型態'].values.tolist()
rex1 = r'(.*)(\(){01}'
for row in list1:
    try:
        match = re.search(rex1, row)
        house_type.append(match[1])
    except:
        house_type.append(row)
df1 = pd.DataFrame(house_type, columns = ['建物類型'])
print(df1.shape)
df1.head()
```

```python
rent7 = rent7.reset_index(drop=True)
df1 = df1.reset_index(drop=True)
rent7 = pd.concat([rent7 , df1], axis = 1)
rent7 = rent7.drop(['建物型態'],axis = 1)
print(rent7.shape)
rent7
```
