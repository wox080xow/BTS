### 臺北市搶救不易狹小巷道清冊

```python
# 巷道寬度在4公尺以下

from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

import time


import os
import re

from bs4 import BeautifulSoup
import pprint
import lxml 


options = webdriver.ChromeOptions()
#options.add_argument("--headless")
options.add_argument("--start-maximized")         # 最大化視窗
options.add_argument("--incognito")               # 開啟無痕模式
options.add_argument("--disable-popup-blocking ") # 禁用彈出攔截

driver = webdriver.Chrome()

driver.get('https://www.google.com/maps/d/u/0/viewer?mid=1etnqfC2T0bnijYA15fqEdRk6UVA&ll=25.024863900000017%2C121.5345558&z=15')
time.sleep(5)

# 有191個項目，這裡取10個test
for num in range (1,11):

    road=driver.find_element(By.CSS_SELECTOR,"#legendPanel > div > div > div.i4ewOd-PBWx0c-bN97Pc-haAclf > div > div > div.i4ewOd-pbTTYe-haAclf > div:nth-child(1) > div > div.HzV7m-pbTTYe-r4nke > label")
    driver.execute_script("arguments[0].click();", road)  # 有擋爬蟲 需要js 語法
    time.sleep(2)
    
    a=f"#legendPanel > div > div > div.i4ewOd-PBWx0c-bN97Pc-haAclf > div > div > div.i4ewOd-pbTTYe-haAclf > div:nth-child(1) > div > div.HzV7m-pbTTYe-bN97Pc > div.HzV7m-pbTTYe-JNdkSc > div.HzV7m-pbTTYe-JNdkSc-PntVL > div:nth-child({num}) > div > div"
    #print(a)       
    road2=driver.find_element(By.CSS_SELECTOR,a)
    driver.execute_script("arguments[0].click();", road2) # 有擋爬蟲 需要js 語法
    time.sleep(2)
    
    elem=driver.find_element_by_xpath("//*")  # 選擇此頁所有 path
    source_code = elem.get_attribute("outerHTML") # get 此頁所有html
    soup = BeautifulSoup(source_code, "lxml") # 指定 lxml 作為解析器
    tag = soup.find_all("div", class_="qqvbed-p83tee-lTBxed")
    #print(type(tag))  
   
    for x in tag: 
        y=str(x)
        
        z =r'lTBxed\"\>(.+)\<'
        matchIterable = re.finditer(z, y)
        list_Road=[]
        for match in matchIterable:
            z=match.group(1)
            #print(z)
            list_Road.append(z)
            list_Road.append('\n')
            
    string_Road= ''.join(list_Road)
    print(string_Road)
    fp = open(f"巷道寬度在4公尺以下.csv", "a",encoding='UTF-8')
    fp.write(string_Road)
    fp.close()
           
    driver.refresh()
    time.sleep(3) 
print('完成了')
    

    
    
```
