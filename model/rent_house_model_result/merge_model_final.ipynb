### 將多元回歸模型結果之建物坪數和用途種類做整理

```python
import numpy as np
import pandas as pd
# 讀取租屋模型預測特徵及租金漲跌資料
rent_house = pd.read_csv('RentHouseRe2_modify.csv', encoding= 'big5')
rent_house
```

```python
# 讀取租屋模型預測資料
rent_house_pre = pd.read_csv('RentAreaSummary_modify.csv', encoding= 'big5')
rent_house_pre
```

```python
# 把 3-10坪建築物單獨取出
rent_house_3_10 = rent_house[(rent_house['建物總坪數']>3)&(rent_house['建物總坪數']<=10)].groupby('Area').count()

df1 = pd.DataFrame({'Area':rent_house_3_10.index, '坪數 3-10':rent_house_3_10['建物總坪數'].values})
df1
```

```python
#由上述，將建物依照坪數分割

rent_house_10_30 =rent_house[(rent_house['建物總坪數']>10)&(rent_house['建物總坪數']<=30)].groupby('Area').count()
df2 = pd.DataFrame({'Area':rent_house_10_30.index, '坪數 10-30':rent_house_10_30['建物總坪數'].values})

rent_house_30_50 = rent_house[(rent_house['建物總坪數']>30)&(rent_house['建物總坪數']<=50)].groupby('Area').count()
df3 = pd.DataFrame({'Area':rent_house_30_50.index, '坪數 30-50':rent_house_30_50['建物總坪數'].values})

rent_house_50_70 = rent_house[(rent_house['建物總坪數']>50)&(rent_house['建物總坪數']<=70)].groupby('Area').count()
df4 = pd.DataFrame({'Area':rent_house_50_70.index, '坪數 50-70':rent_house_50_70['建物總坪數'].values})

rent_house_70_90 = rent_house[(rent_house['建物總坪數']>70)&(rent_house['建物總坪數']<=90)].groupby('Area').count()
df5 = pd.DataFrame({'Area':rent_house_70_90.index, '坪數 70-90':rent_house_70_90['建物總坪數'].values})
```

```python
#由上述方法，將建物主要用途分出國民住宅和住家用之數量
rent_house_classification =rent_house[rent_house['主要用途'].str.contains('國民住宅')].groupby('Area').count()
df6 = pd.DataFrame({'Area':rent_house_classification.index, '國民住宅':rent_house_classification['主要用途'].values})

rent_house_classification_2 =rent_house[rent_house['主要用途'].str.contains('住家用')].groupby('Area').count()
df7 = pd.DataFrame({'Area':rent_house_classification_2.index, '住家用':rent_house_classification_2['主要用途'].values})
```

```python
# 將建物坪數分割合併成表
result = pd.concat([rent_house_pre, df1['坪數 3-10'], df2['坪數 10-30'], df3['坪數 30-50'], df4['坪數 50-70'], df5['坪數 70-90'], df6['國民住宅'], df7['住家用']], axis=1)
result
```

```python
# 填滿空值
result2=result.fillna(0)
# 去掉不要的欄位
result3=result2.drop(['Unnamed: 0'], axis=1)
result3
```

```python
# 寫出csv
result3.to_csv('House_style_modify.csv', index=None)
```

```python
# 將交易量分割出來
rent_house_count=pd.concat([rent_house_pre['Area'], rent_house_pre['count']], axis=1)

# 將交易量合併至特徵表
rent_house_count2=rent_house.merge(rent_house_count, on = 'Area')

# 去掉不要的欄位
rent_house_count3=rent_house_count2.drop(['Unnamed: 0'], axis=1)
rent_house_count3
```

```python
# 寫出csv
rent_house_count3.to_csv('RentHouseRe2_count_modify.csv', index=None)
```
